## Load the data

In this section we load the data and display it in a interactive viewer

```python
input_image_path1 = './mv3blebbing/1_CAM01_000004.tif'
input_image_path2 = './mv3blebbing/1_CAM01_000008.tif'
print('Image path loaded')
```

## Interactive parameters selection

```python
output_image = None
filepath1 = None
filepath2 = None
filepath3 = None
input_basename3 = None

def run_flowscope(gamma, maximum_iteration):
    global output_image
    global filepath1
    global filepath2
    global filepath3
    global input_basename3
    
    input_basename1 = ntpath.basename(input_image_path1)
    input_basename2 = ntpath.basename(input_image_path2)
    input_basename3 = (ntpath.basename(input_image_path1) + "_" + ntpath.basename(input_image_path2))
    
    myprocess = BiProcess('./flowscope-wrapping/wrappers/flowscope.xml')
    output_image = myprocess.exec('-s', input_image_path1,
                '-t', input_image_path2,
                '-g', gamma,
                '-i', maximum_iteration,
                '-o', './Output')

    clear_output()
    print('Job done. Your output images are in the "Output" folder')

interact_manual(run_flowscope, gamma = (0, 2.4), maximum_iteration = (0,12))
```

```python
# Type your x, y and z values
x = 74
y = 69
z = 53
```

## Flow Visualization

```python
# XY

filepath1 = "./Output/" + input_basename3 + "soru.tif"
filepath2 = "./Output/" + input_basename3 + "sorv.tif"
filepath3 = "./Output/" + input_basename3 + "sorw.tif"


u = io.imread(filepath1)
v = io.imread(filepath2)
w = io.imread(filepath3)


U = 1*u[z, :, :]
#print(U)

V = 1*v[z, :, :]
#print(V)

flo = np.stack((U, V), axis = 2) 
print(flo.shape)
flows = np.stack((flo, flo), axis = 0) 
print(flows.shape)

colors = colorflow.motion_to_color(flows)
flowanimXY = animate.FlowAnimation(colors, scale=1)
HTML(flowanimXY.to_jshtml())
```

```python
# XZ

U = 1*u[y, :, :]
print("U = {}".format(U), "\n")
U = np.transpose(U)

W = 1*w[y, :, :]
print("W = {}".format(W), "\n")
W = np.transpose(W)

flo = np.stack((U, W), axis = 2) 
print("flo.sahpe = {}".format(flo.shape))

flows = np.stack((flo, flo), axis = 0) 
print("flows.shape = {}".format(flows.shape))

colors = colorflow.motion_to_color(flows)
flowanimXZ = animate.FlowAnimation(colors, scale=1)
HTML(flowanimXZ.to_jshtml())
```

```python
# YZ

V = 1*v[:, x, :]
print("V = {}".format(V), "\n")
#V = np.transpose(V)

W = 1*w[:, x, :]
print("W = {}".format(W), "\n")
#W = np.transpose(W)

flo = np.stack((W, V), axis = 2) 
print("flo.sahpe = {}".format(flo.shape))

flows = np.stack((flo, flo), axis = 0) 
print("flows.shape = {}".format(flows.shape))

colors = colorflow.motion_to_color(flows)
flowanimYZ = animate.FlowAnimation(colors, scale=1)
HTML(flowanimYZ.to_jshtml())
```

## Compute Magnitude

```python
mag = (u**2 + v**2 + w**2)
mag = np.sqrt(mag)
io.imsave("./Output/mag.tif", mag) #the 3D magnitude is saved here
```
